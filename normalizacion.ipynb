```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from dataset import load_dataset
from util import apply_target, targetNormalize
```

```python
def print_features(input, title):
    x = np.arange(len(input))
    for i in range(10):
        plt.scatter(x, input[:,i], alpha=0.3)
        
    plt.xlabel('instancias')
    plt.ylabel('valor de los parmetros')
    plt.title(title)
    plt.show()

def showNormalization(ds_file, input_break, target_break, apply_target):
    df = pd.read_csv(ds_file, header=None)

    data_input = df[df.columns[
        input_break[0]:input_break[1]]]
    
    data_target = df[df.columns[
        target_break[0]:target_break[1]]
    ]

    normalized_input = targetNormalize(data_input)
    input = data_input.to_numpy()
    target = apply_target(data_target)

    print_features(input, 'sin normalizar')
    print_features(normalized_input, 'normalizado')
    
    x = np.arange(len(input))

    fig, ax1 = plt.subplots()

    plt.xlabel('instancias')
    plt.ylabel('valor de los parmetros')
    plt.title('propiedad 8')

    ax1.set_xlabel('instancias')
    ax1.set_ylabel('valor de los parmetros')
    ax1.scatter(x, input[:,7], label='sin normalizar', s=[70 for i in x], alpha=0.7)
    ax1.tick_params(axis='y', labelcolor='black')

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    ax2.scatter(x, normalized_input[:,7], label='normalizado', color='orange', s=[25 for i in x], alpha=1)
    ax2.tick_params(axis='y', labelcolor='black')

    fig.legend()
    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()

    
showNormalization('./dataset/tp1_ej1_training.csv', (1, 11), (0, 1), apply_target)
```
